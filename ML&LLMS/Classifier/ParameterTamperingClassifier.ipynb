```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('../datasets/parameterTampering/pt_dataset.json', orient='records') 
```

```python
counts = df['label'].value_counts()
```

```python
counts.plot.pie(figsize=(5, 5))
```

```python
X = df['length'].to_numpy().astype(str)
y = df['label'].to_numpy().astype(str)
```

```python
print(len(X))
print(len(y))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
cross_val_score(clf, X.reshape(-1, 1), y, cv=10)
```

```python

clf.fit(X.reshape(-1, 1), y)
```

```python
import joblib

joblib.dump(clf, 'pt_predictor.joblib')
```
